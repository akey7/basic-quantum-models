```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin, sqrt, pi
```

```python
class ParticleInABox1d:
    def __init__(self, mass, length):
        self.mass = mass
        self.length = length
        
    def wavefunction(self, x, n):
        return sqrt(2 / self.length) * sin(n * pi * x / self.length)
        
    def sq_wavefunction(self, x, n):
        return self.wavefunction(x, n) ** 2
```

```python
box = ParticleInABox1d(1, 1)
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
fontdict = {'fontsize': 10}
xs = np.linspace(0, 1, 100)

ys1 = [box.sq_wavefunction(x, n=1) for x in xs]
axs[0, 0].set_ylabel('1(x)^2')
axs[0, 0].set_xlabel('x')
axs[0, 0].plot(xs, ys1)

ys2 = [box.sq_wavefunction(x, n=2) for x in xs]
axs[0, 1].set_title('n = 2')
axs[0, 1].set_ylabel('2(x)^2')
axs[0, 1].set_xlabel('x')
axs[0, 1].plot(xs, ys2)

ys3 = [box.sq_wavefunction(x, n=3) for x in xs]
axs[1, 0].set_title('n = 3')
axs[1, 0].set_ylabel('3(x)^2')
axs[1, 0].set_xlabel('x')
axs[1, 0].plot(xs, ys3)

ys4 = [box.sq_wavefunction(x, n=4) for x in xs]
axs[1, 1].set_title('n = 4')
axs[1, 1].set_ylabel('4(x)^2', fontsize=10)
axs[1, 1].set_xlabel('x')
axs[1, 1].plot(xs, ys4)
```
